# function for cleaning tweets

```python
def getCleanTwit(str1):
    words= str1.split()
    clean_words = []
    for i in words:
        if re.findall(r'^http', i):
            continue
        else:
            clean_words.append(i)
    text = u' '.join(x for x in clean_words)
    text = re.sub("http*:\/\/[0-9a-zA-Z\-\_]+\.[\-\_0-9a-zA-Z]", u"", text)
    text = re.sub("\@[\_0-9a-zA-Z]+\:?", u"", text)
    text = re.sub("\&amp\;?", u"", text)
    text = re.sub("[\:\.]{1,}$", u"", text)
    text = re.sub("^RT\:?", u"", text)

    return text
```

```python
from twitterscraper.query import query_user_info
import pandas as pd

locations = ['NA']* len(username)

for i in range(len(username)):
    try:
        user_info = query_user_info(username[i])   
    except:
        pass
    try:
        location = user_info.location
        if location:
            locations[i] = location
    except (UnboundLocalError, AttributeError):
        pass
    
```

```python
from geopy.geocoders import Nominatim
import gmplot

geolocator = Nominatim()

coor_lat = ['NA']* len(locations)
coor_lon = ['NA']* len(locations)


# Go through all tweets and add locations to 'coordinates' dictionary
coordinates = {'latitude': [], 'longitude': []}
for i in range(len(locations)):
    if locations[i] == 'NA':
        pass
    try:
        location = geolocator.geocode(locations[i])
        
        # If coordinates are found for location
        if location:
            coor_lat[i] = location.latitude
            coor_lon[i] = location.longitude
            
    # If too many connection requests
    except:
        pass
    
```
